```python
#Python Initiation
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import sympy
from scipy.interpolate import make_interp_spline, BSpline
%matplotlib inline

#Graph Formatting
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 14
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
Res=300 #Resoultion of the calulations

#Editing Font Size for Plot
font = {'size'   : 18}
matplotlib.rc('font', **font)
```

## Plot of Figure 3.5 from Rocket Propulsion Elements

```python
# Figure 3.5
# Inputs
k_val=[1.2]
ε=50
P1_P2=np.linspace(10,10000,1000)
P2_P1=1/P1_P2

#Primary Line Equation + Plot
for k in k_val: 
    CF=np.sqrt((((2*k**2)/(k-1))*(2/(k+1))**((k+1)/(k-1))*(1-(P2_P1)**((k-1)/k))))
    plt.plot(P1_P2,CF)
    
#Equations to be Updated Later
#ε=((k+1)/2)**(1/(k-1))*(P2_P1)**(1/k)*((k+1)/(k-1)*(1-P2_P1**((k-1)/k)))**(1/2)
CF_=k**3+P2_P1*ε

#Plotting Parameters
plt.ylabel('$C_F$ (Unitless)')
plt.xlabel('$P_1/P_2$ (Unitless)')
plt.plot(P1_P2,CF_)
plt.xscale('log')
plt.xlim(10,10000)
plt.ylim(1.2,2.3)
plt.legend(['k=1.2','ε=50'])
plt.show()
```
